```python
import numpy as np
import pandas as pd
```

```python
cols = ['Age','Workclass',"Fnlwgt","Education","Education-num","Marital_status","Occupation","Relationship","Race","Sex","Capital_gain","Capital_loss","Hours_per_week","Native_Country","Salary"]
```

```python
df = pd.read_csv("adult.csv", header=None)
df.columns = cols
df.head()
```

# Subsets

```python
df["Native_Country"].unique()
```

```python
df[df == " ?"].count()
```

```python
df.replace(" ?", pd.NA, inplace=True)
df = df.dropna()
df
```

```python
df.isna().sum()
```

```python
dfs1 = df[df["Sex"] == " Male"]
dfs1
```

```python
dfs2 = df[df["Sex"] == " Female"]
dfs2
```

```python
dfs3 = df[df["Race"] == " Black"]
dfs3
```

# Merging

```python
df_merge = pd.merge(dfs1, dfs3, how="right")
df_merge
```

# Sorting

```python
df_sort = df.sort_values(by="Age", ascending=False)
df_sort
```

# Transpose

```python
df_transpose = df.T
df_transpose
```

# Melting to Long Format

```python
df_melt = pd.melt(df, id_vars=["Age","Workclass","Race","Sex"])
df_melt
```

# Casting to Wide Format

```python
df_cast = pd.pivot_table(df, index=["Sex","Workclass","Race"], values=["Fnlwgt","Capital_gain","Capital_loss","Hours_per_week"], aggfunc="mean", fill_value= 0)
df_cast
```
