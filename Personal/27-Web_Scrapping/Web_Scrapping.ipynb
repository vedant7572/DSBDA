```python
import requests
from bs4 import BeautifulSoup as bs
import numpy as np
import pandas as pd
```

```python
request = requests.get("https://www.flipkart.com/hp-victus-intel-core-i5-12th-gen-12450h-8-gb-512-gb-ssd-windows-11-home-4-graphics-nvidia-geforce-rtx-2050-50-tgp-15-fa1132tx-gaming-laptop/product-reviews/itma3a4c675cefe2?pid=COMGSV4RCCMG5FJ8&lid=LSTCOMGSV4RCCMG5FJ8UF3HY5&marketplace=FLIPKART")
request
```

```python
soup = bs(request.content, 'html.parser')
print(soup.prettify())
```

```python
names = soup.find_all('p', class_='_2NsDsF AwS1CA')
names
```

```python
cust_names = []
for i in range(0, len(names)):
    cust_names.append(names[i].get_text())
cust_names
```

```python
titles = soup.find_all('p', class_='z9E0IG')
titles
```

```python
review_titles = []
for i in range(0,len(titles)):
    review_titles.append(titles[i].get_text())
review_titles
```

```python
ratings = soup.find_all(class_=['XQDdHH Ga3i8K','XQDdHH Js30Fc Ga3i8K'])
ratings
```

```python
review_ratings = []
for i in range(0,len(ratings)):
    review_ratings.append(ratings[i].get_text())
review_ratings
```

```python
descp = soup.find_all(class_='ZmyHeo')
descp
```

```python
review_descp = []
for i in range(0,len(descp)):
    review_descp.append(descp[i].get_text().replace('READ MORE',"").strip())
review_descp
```

```python
df = pd.DataFrame()
df['Customer Name'] = cust_names
df['Review Title'] = review_titles
df['Rating'] = review_ratings
df['Review Description'] = review_descp
```
