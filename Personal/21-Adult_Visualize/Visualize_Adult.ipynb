```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore")
```

```python
df = pd.read_csv("adult.csv", header=None)
df.head()
```

```python
cols = ['Age','Workclass',"Fnlwgt","Education","Education-num","Marital_status","Occupation","Relationship","Race","Sex","Capital_gain","Capital_loss","Hours_per_week","Native_Country","Salary"]
df.columns = cols
df.head()
```

```python
df.shape
```

```python
df.info()
```

# Data Cleaning

```python
df[df == " ?"].count()
```

```python
df.head()
```

```python
df_q = df.replace(" ?", np.nan)
df_q.isna().sum()
```

```python
df = df.fillna(df.mode())
df.head()
```

```python
df.isna().sum()
```

# Plots

## Histogram

```python
plt.hist(df['Age'], bins=10, color='Blue', edgecolor='Black')
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.title("Histogram of Age")
plt.show()
```

## Dot Plot

```python
plt.plot(df['Capital_gain'], linestyle='', marker='o')
plt.xlabel('Index')
plt.ylabel("Capital Gain")
plt.title("Dot Plot of Capital Gain")
plt.show()
```

```python
plt.bar(df['Salary'].unique(),df['Salary'].value_counts(), color='Orange')
plt.xlabel("Salary")
plt.ylabel("Frequency")
plt.title("Bar Plot of Salary")
plt.show()
```

```python
# plt.figure(figsize=(10,8))
plt.plot(df['Hours_per_week'], linestyle='-', color='purple')
```

```python
size = df['Education'].value_counts()
labels = df['Education'].unique()
plt.figure(figsize=(10,8))
plt.pie(size, labels=labels, autopct='%1.1f%%')
```

```python
plt.boxplot(df['Age'], notch=True, patch_artist=True)
```

```python
sns.scatterplot(x= 'Age', y= 'Fnlwgt', data=df, hue='Salary')
```

```python
sns.pairplot(df[['Age','Fnlwgt','Capital_gain','Hours_per_week','Salary']], hue='Salary', diag_kind='hist', kind='scatter')
```
