# Create Subsets

```python
NotAffected = df[df["area"] == 0]
NotAffected
```

```python
PartiallyAffected = df[(df["area"] > 0) & (df["area"] <= 13)]
PartiallyAffected
```

```python
FullyAffected = df[df["area"] > 13]
FullyAffected
```

# Merge Subset

```python
df_merged = pd.merge(NotAffected, PartiallyAffected, how="outer")
df_merged
```

# Sort

```python
df_sort_temp = df.sort_values(by="temp", ascending=False)
df_sort_temp
```

```python
df_sort_wind = df.sort_values(by="wind", ascending=False)
df_sort_wind
```

```python
df_sort_area = df.sort_values(by="area", ascending=False)
df_sort_area
```

# Transpose

```python
df_transpose = df.T
df_transpose
```

# Melt

```python
df_melt = pd.melt(df, id_vars = ["month","day"])
df_melt
```

# Cast

```python
df_cast = pd.pivot_table(df, index=["day","month"], values=["DC","area"])
df_cast
```
