# Subsets

```python
df_type_photo = df[df["Type"]=="Photo"]
df_type_photo
```

```python
df_type_link = df[df["Type"] == "Link"]
df_type_link
```

```python
df_type_link.shape
```

# Merge subsets

```python
df_merged = pd.concat([df_type_photo, df_type_link])
df_merged
```

```python
df_s1 = df.iloc[0:100, 1:5]
df_s1
```

```python
df_s2 = df.iloc[0:100, 4:10]
df_s2
```

```python
df_merge = pd.merge(df_s1, df_s2, how="outer")
df_merge
```

# Sort

```python
df_sort = df.sort_values(by = ["like"],ascending=False)
df_sort
```

```python
df_sort_share = df.sort_values(by=["share"], ascending=True)
df_sort_share
```

```python
df_melt = pd.melt(df, id_vars=["Type","Category"], var_name="All", value_name="Value")
df_melt
```

```python
df['index'] = range(1, len(df)+1)
df.head()
```

```python
df_melt1 = pd.melt(df, id_vars=['index','Type','Category','Paid'], value_vars=["comment","like","share"], var_name= "Metrics")
df_melt1
```

```python
df[df["index"] == 500].count()
```

```python
df_wide = pd.pivot_table(df_melt1, index=["Type","Category"], columns="Metrics", values="value", aggfunc="mean", fill_value=0)
df_wide
```
