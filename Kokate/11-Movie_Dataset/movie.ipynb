# Melt

```python
df
```

```python
df['genres'].unique()
```

```python
cols = ['revenue', 'vote_average', 'vote_count']
df[cols] = df[cols].astype(float)
```

```python
melted = pd.melt(df, id_vars=["adult", 'genres'], value_vars=['revenue', 'vote_average', 'vote_count'], value_name="Value", var_name="Metrics")
melted
```

# Casting

```python
df_casted = pd.pivot_table(melted, index=["adult", 'genres'], columns="Metrics", values="Value", aggfunc="mean", fill_value=0)
df_casted
```
