```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("Iris.csv", header=None)
df.head()
```

```python
df.columns = ['SepalLengthCm',"SepalWidthCm", "PetalLengthCm", "PetalWidthCm", "Species"]
df.head()
```

```python
df.info()
```

```python
df.shape
```

# Create Subsets

```python
df['Species'].unique()
```

```python
df_setosa = df[df["Species"] == 'Iris-setosa']
print(df_setosa.shape)
df_setosa.head(10)
```

```python
df_versi = df[df["Species"] == "Iris-versicolor"]
print(df_versi.shape)
df_versi.head()
```

# Merging

```python
df_merged = pd.concat([df_setosa, df_versi])
df_merged
```

```python
dfs1 = df.iloc[0:50, 0]
dfs1.head()
```

```python
dfs2 = df.iloc[50:100, 0:2]
dfs2.head()
```

```python
df_merge = pd.concat([dfs1, dfs2])
df_merge
```

```python
df_merge1 = pd.merge(dfs1, dfs2, on="SepalLengthCm", how="right")
df_merge1
```

```python
df_merge1.isna().sum()
```

# Sort Values

```python
df_sorted = df.sort_values(by="PetalLengthCm", ascending=True)
df_sorted
```

# Transpose

```python
df_transpose = df.transpose();
df_transpose
```

# Melting to long format

```python
df['Index'] = range(1, len(df)+1)
df.head()
```

```python
df_melt = pd.melt(df, id_vars="Index") //value_vars used to specify the specific variable we want.
df_melt
```

# Cast to Wide Format

```python
df_cast = pd.pivot_table(df, index=["Species","PetalLengthCm"], aggfunc="mean", fill_value= 0)
df_cast
```
