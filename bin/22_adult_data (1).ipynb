```python
df=pd.read_csv('adult.csv',header=None)
df.head()
```

```python
df.columns=['age','workclass','fnlwgt','education','education-num	','marital-status	','occupation','relationship','race','sex','capital-gain','capital-loss','hours-per-week','native-country','Target']
df.head()
```

```python
df.isna().sum().sum()
```

```python
plt.figure(figsize=(8,6))
plt.hist(df['age'],color='skyblue',edgecolor='black')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(8,6))
df['Target'].value_counts().plot(kind='pie')
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.boxplot(x='Target',y='education',data=df)
plt.show()
```

```python
plt.figure(figsize=(12,8))
sns.scatterplot(x='workclass',y='hours-per-week',hue='sex',style='marital-status\t',data=df)
plt.xlabel('Work Class')
plt.ylabel('Working hours per week')
plt.show()
```
