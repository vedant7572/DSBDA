```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('heart.csv')
df.head()
```

```python
df.isnull().sum()
```

```python
(df.values=='?').sum()
```

```python
df1=df[df>=0]
df1.head()
```

```python
plt.figure(figsize=(8,6))
plt.hist(df['age'],bins=10,color='skyblue', edgecolor='black')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.title("Age Distribution")
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.stripplot(x='cp',y='chol',jitter=True,data=df,hue='sex')
plt.xlabel('CP')
plt.ylabel('Cholestro')
plt.title('Cholestrol level as per age')
plt.show()
```

```python
# plt.figure(figsize=(8,6))
# sns.barplot(x='age',y='chol',data=df)
# plt.xlabel('age')
# plt.ylabel("Trestbps")
# plt.title('Age VS Trsetbps')
# plt.show()
plt.bar(df['age'],df['chol'])
plt.xlabel('Age')
plt.ylabel('Cholestrol')
plt.show()
```

```python
plt.figure(figsize=(8,6))
plt.plot(df['thalach'])
plt.xlabel('Index')
plt.ylabel('Heart Rate')
plt.show()
```

```python
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
sns.boxplot(x='target', y='chol', data=df)
plt.title('Box Plot of Cholesterol by Heart Disease')

plt.subplot(1, 2, 2)
sns.scatterplot(x='chol', y='trestbps', data=df, hue='target')
plt.title('Scatter plot of Cholesterol vs Resting Blood Pressure with Heart Disease')

plt.show()
```

```python
plt.figure(figsize=(8,6))
df['target'].value_counts().plot(kind='pie',autopct='%1.1f%%')
plt.title("Target Distribution")
plt.show()
```

```python
plt.figure(figsize=(8,6))
df['sex'].value_counts().plot(kind='pie',autopct='%1.1f%%',colors=['blue','black'])
plt.show()
```
