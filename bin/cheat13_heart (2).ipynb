```python
import pandas as pd
import numpy as np
from scipy import stats
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score,classification_report
```

```python
dataframe=pd.read_csv("heart.csv")
dataframe.head()
```

```python
dataframe=dataframe.replace('?',pd.NA)
```

```python
dataframe.shape
```

```python
dataframe1=dataframe.dropna()
```

```python
dataframe.isnull().sum()
```

```python
dataframe=dataframe1[dataframe1>=0]
dataframe.head()
```

```python
dataframe.shape
```

```python
# def clean(df):
#   for col in df:

#     df[col].fillna(df[col].mean()  , inplace = true)

# clean(dataframe)
```

```python
z_index=np.abs(stats.zscore(dataframe))
threshold=3
clean_data=dataframe[(z_index<threshold).all(axis=1)]
clean_data.shape
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
clean_data.loc[clean_data['sex']==1,'sex']='M'
clean_data.loc[clean_data['sex']==0,'sex']='F'
lb=LabelEncoder()
clean_data['sex']=lb.fit_transform(clean_data['sex'])
clean_data
```

```python
X=clean_data.drop('target',axis=1)
Y=clean_data['target']
```

```python
x_train,x_test,y_train,y_test=train_test_split(X,Y,test_size=0.2,random_state=46)
```

```python
x_test.shape
```

```python
rm=LogisticRegression()
rm.fit(x_train,y_train)
```

```python
rm_result=rm.predict(x_test)
print(rm.predict(x_test))
```

```python
km=KNeighborsClassifier()
km.fit(x_train,y_train)
```

```python
km_result=km.predict(x_test)
print(km.predict(x_test))
```

```python
print("Accuracy Sore for logistic regression:",accuracy_score(y_test,rm_result))
```

```python
print("Accuracy Sore for KM model:",accuracy_score(y_test,km_result))
```

```python
print(classification_report(y_test,rm_result))
print(classification_report(y_test,km_result))
```
