```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('adult.csv',header=None)
df.head()
```

```python
df.columns=['age','workclass','fnlwgt','education','education-num','marital-status','occupation','relationship','race','sex','capital-gain','capital-loss','hours-per-week','native-country','target']
```

```python
df.isna().sum().sum()
```

```python
(df.values=='?').sum()
```

```python
df.head()
```

```python
plt.figure(figsize=(8,6))
plt.hist(df['age'],color='skyblue',edgecolor='black')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(8,16))
sns.stripplot(x='education',y='hours-per-week',hue='sex',data=df,jitter=True)
plt.show()
```

```python
plt.figure(figsize=(8,16))
plt.bar(df['education'],df['capital-gain'])
plt.xlabel('Eductaion')
plt.ylabel('Capital')
plt.show()
```

```python
plt.figure(figsize=(8,16))
sns.scatterplot(x='education-num',y='hours-per-week',data=df,hue='sex',style='target')
plt.show()
```
