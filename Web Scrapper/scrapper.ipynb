```python
import requests
import bs4
```

```python
x=requests.get("https://www.flipkart.com/nothing-phone-1-black-128-gb/product-reviews/itmeea53a564de47?pid=MOBGCYGPFEGDMYQR&lid=LSTMOBGCYGPFEGDMYQROW9GCT&marketplace=FLIPKART")
```

```python
x
```

```python
x.content
```

```python
soup=bs4.BeautifulSoup(x.text)
print(soup.prettify())
```

```python
reviews_content=[]
reviews=soup.findAll('div',{'class':'ZmyHeo'})
for i in range(0,len(reviews)):
    reviews_content.append(reviews[i].get_text())
    
reviews_content
```

```python
overall=soup.find('div',{'class':'ipqd2A'}).get_text()

overall
```

```python
ratings=[]
individual_rate=soup.findAll('div',{'class':'XQDdHH Ga3i8K'})
for i in range(0,len(individual_rate)):
    ratings.append(individual_rate[i].get_text())
    
    
ratings
```

```python
title=[]
review_tit=soup.findAll('p',{'class':'z9E0IG'})
for i in range(0,len(review_tit)):
    title.append(review_tit[i].get_text())
    
title
```

```python
names=[]
cust_name=soup.findAll('p',{'class':'_2NsDsF AwS1CA'})
for i in range(0,len(cust_name)):
    names.append(cust_name[i].get_text())
    
names
```

```python
import pandas as pd
df=pd.DataFrame()
df['customerName']=names[:5]
df['reviewTitle']=title[:5]
df['customerRatings']=ratings[:5]
df['reviewsContent']=reviews_content[:5]
```

```python
df
```

```python
df.to_csv(r'./reviews.csv',index=True)
```
